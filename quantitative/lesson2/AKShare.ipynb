# AKShare Tutorial
Notebook to show how to use the AKShare library to get financial data for Big A.

```python
import akshare as ak
import pandas as pd
import numpy as np
import mplfinance as mpl
```

# 历史行情数据测时
## 日频数据

```python
# 沪深京A
stock_zh_a_spot_df = ak.stock_zh_a_spot_em()
stock_zh_a_spot_df
```

```python
sec = stock_zh_a_spot_df[stock_zh_a_spot_df['名称']=='中视传媒']
sec
```

```python
sec_code = sec['代码'].values[0]
sec_code
```

```python
price = ak.stock_zh_a_hist(sec_code, period="daily", start_date="20231201", end_date="20240219")
price
```

```python
price['日期']
```

```python
# 取 price 的 Open，High，Low，Close，Volume 列
price = price[['开盘', '最高', '最低', '收盘', '成交量', '日期']]
price.columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Date'] # rename columns
# set price '日期' column as index with datetime format YYYY-MM-DD
price['Date'] = pd.to_datetime(price['Date'])
price.set_index('Date', inplace=True)
price
```

```python
mpl.plot(price, type='candle', volume=True, style='yahoo', mav=(3, 10, 20), figsize=(20, 10))
```
