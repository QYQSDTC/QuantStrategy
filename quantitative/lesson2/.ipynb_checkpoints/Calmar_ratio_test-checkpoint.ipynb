用卡玛比率来寻找强势股

```python
stock_pool = jq.get_all_securities(types=['stock'], date=None)
stock_codes = stock_pool.index.to_list()
```

```python
prices = jq.get_price(security=stock_codes, count=300,end_date=today, frequency='daily', skip_paused=True)
prices.head(10)
```

```python
fin_prices = pd.read_csv('A_stock_prices.csv', index_col=0)
fin_prices.head(10)
```

```python
# calculate Calmar ratio
def calmar(s):
    ret=(s/s.iloc[0]-1).iloc[-1]
    drawdown=(1-s/s.cummax()).max()
    kama=ret/drawdown
    return kama

# 计算250日期间收益
data = fin_prices[-250:]
rets = data.apply(lambda s:(s/s[0] - 1)[-1])
drawdowns = data.apply(lambda s:(1-s/s.cummax()).max())
calmars = data.apply(lambda s:calmar(s))
calmar_df=pd.concat([rets,drawdowns,calmars],axis=1).rename({'0':'rets','1':'Drawdowns', '2': 'Calmars'})
calmar_df
```
