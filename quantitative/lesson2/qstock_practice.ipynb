# Calculate MAX using TALib

```python
# get prices for 杭州热电
data = qs.get_data('杭州热电', start='20230101', freq=120)
data.tail()
```

```python
MA3 = ta.MA(data['close'], timeperiod=3)
data['MA3'] = MA3
data.tail()
```

```python
MA10 = ta.MA(data['close'], timeperiod=10)
data['MA10'] = MA10
data.tail()
```
